```python
from ipyleaflet import Map, MagnifyingGlass, basemaps, basemap_to_tiles, Marker
import pandas as pd
```

```python
# lee el data json
datos = pd.read_json("data/country-codes-lat-long-alpha3.json")
```

```python
# pasa a lista

lista = datos['ref_country_codes'].tolist()
```

```python
# presenta lista
lista
```

```python
#presenta tal cual cada JSON
datos
```

```python
#Crea un DataFrame de la lista 
datos_paises = pd.DataFrame(lista)
```

```python
datos_paises
```

```python
# sacar de lista los paises que se referencian 
datos_valores = datos_paises.to_dict("records")

datos_valores[0]

datos_valores = [d for d in datos_valores if d['country'] in ["Ecuador", "Colombia", "Chile", "Japon"\
                                                             "Venezuela, Bolivarian Republic of"]]
datos_valores
```

```python
#punto de partida
center = (-34, -64)

m = Map(center=center, zoom=2)
# agrega puntos al mapa 
for d in datos_valores:
    #genera una tupla de informacion 
    c = (d['latitude'], d['longitude'])
    
    marker = Marker(location=c, draggable=False)
    #agrega al mapa cada marca
    m.add_layer(marker);


# m.add_layer(magnifying_glass)
m
```

```python
center = (-34, -64)
m = Map(center=center, zoom=2)
for d in datos_valores:
    c = (d['latitude'], d['longitude'])
    print(d['country'])
    marker = Marker(location=c, draggable=False, title=d['country'], alt=d['country'])
    m.add_layer(marker);


# m.add_layer(magnifying_glass)
m
```
