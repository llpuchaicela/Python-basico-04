```python
from ipyleaflet import Map, MagnifyingGlass, basemaps, basemap_to_tiles, Marker
import pandas as pd
```

```python
# lee la data 
datos = pd.read_csv("data/ciudades.csv")
datos
```

```python
    # sacar de lista los paises que se referencian 
datos_valores = datos.to_dict("records")

datos_valores
```

```python
#punto de partida
center = (-1.7929665, -78.1368875)

m = Map(center=center, zoom=6)
# agrega puntos al mapa 
for d in datos_valores:
    #genera una tupla de informacion 
    c = (d['lat'], d['lng'])
    
    marker = Marker(location=c, draggable=False)
    #    marker = Marker(location=c, draggable=False, title="Ciudad: %s\nPaís:%s\n%d" % (d['city'], d['country'], d['population']))
    #agrega al mapa cada marca
    m.add_layer(marker);


# m.add_layer(magnifying_glass)
m
```
