```python
import pandas as pd
import ipywidgets as widgets
```

```python
datos_candidatos = pd.read_spss("data/candidatos2019.sav")
datos_candidatos.shape
```

```python
datos_candidatos
```

```python
datos_candidatos_2 = datos_candidatos[['CANDIDATO_CODIGO', 'CANDIDATO_NOMBRE', 'OP_CODIGO']]
datos_candidatos_2.shape
```

```python
datos_prefectos = pd.merge(datos_prefectos, datos_candidatos_2, on="CANDIDATO_CODIGO")
```

```python
organizaciones_polticas = pd.read_spss("data/organizacionespolticas2019.sav")
```

```python
organizaciones_polticas.keys()
```

```python
datos_prefectos= pd.merge(datos_prefectos,organizaciones_polticas, on= "OP_CODIGO")
```

```python
datos_prefectos.head()
```

```python
#creo lista para busqueda por provincias
opcionesProv =[(d, d) for d in datos_prefectos.PROVINCIA_NOMBRE.unique()]
opcionesProv.sort() #ordenados alfabeticamente
opcionesProv
```

```python
#opcion de provincias
provincia = widgets.Dropdown(options= opcionesProv,  description = 'Provincia')

#boton de opciones
provincia
```

```python
# extraigo datos de la provincia seleccionados
datos_x_provincia = datos_prefectos[datos_prefectos['PROVINCIA_NOMBRE']==provincia.value]
#presento datos por provincia seleccionada
datos_x_provincia
```

```python
datos_x_provincia.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values().plot(kind="barh", width=0.8)
```

```python
datos_x_provincia.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values(ascending=False).plot(kind="bar")
```
