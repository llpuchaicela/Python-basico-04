```python
import pandas  as pd
import ipywidgets as widgets
```

```python
datos_prefectos= pd.read_spss("data/resultadosdignidadprefecto2019.sav")
datos_prefectos
```

```python
datos_candidatos= pd.read_spss("data/candidatos2019.sav")
datos_candidatos.shape
```

```python
datos_candidatos
```

```python
datos_candidatos_2 = datos_candidatos[['CANDIDATO_CODIGO', 'CANDIDATO_NOMBRE', 'OP_CODIGO']]
datos_candidatos_2.shape
```

```python
datos_prefectos = pd.merge(datos_prefectos, datos_candidatos_2, on = 'CANDIDATO_CODIGO')
```

```python
datos_prefectos
```

```python
opciones =[(d,d)for d in datos_prefectos.CANTON_NOMBRE.unique()]
opciones.sort()
```

```python
print(opciones)
```

```python
datos_candidatos = pd.read_spss("data/candidatos2019.sav")
datos_candidatos.shape
```

```python
datos_candidatos.head()
```

```python
canton = widgets.Dropdown(
    options = opciones,
    value = "LOJA",
    description= 'Cantón',
)
canton
```

```python
datos_x_canton= datos_prefectos[datos_prefectos['CANTON_NOMBRE']==canton.value]
datos_x_canton.head()
```

```python
datos_x_canton.groupby('CANDIDATO_NOMBRE')['VOTOS'].sum().sort_values().plot(kind="barh", width = 0.8)
```

```python
datos_x_canton.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values(ascending=False).plot(kind="barh", width = 0.5)
```
