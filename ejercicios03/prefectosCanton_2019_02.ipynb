```python
import pandas  as pd
import ipywidgets as widgets
```

```python
#generar el proceso para prefectos
# alamacenar la data en la variable datos_prefectos y presento los datos
datos_prefectos= pd.read_spss("data/resultadosdignidadprefecto2019.sav")
datos_prefectos
```

```python
#almaceno la data en variable datos_candidatos y presento la data 
datos_candidatos= pd.read_spss("data/candidatos2019.sav")
datos_candidatos
```

```python
# almaceno datos clave en variable datos_candidatos_2 y presento 
datos_candidatos_2 = datos_candidatos[['CANDIDATO_CODIGO', 'CANDIDATO_NOMBRE', 'OP_CODIGO']]
datos_candidatos_2
```

```python
# Almaceno en una variable lo extraido mediante un merge la union de datos de los prefectos y los candidatos
datos_prefectos = pd.merge(datos_prefectos, datos_candidatos_2, on = 'CANDIDATO_CODIGO')
```

```python
# presento 
datos_prefectos
```

```python
#creacion del csv modificado
datos_prefectos.to_csv("data/prefectos_2019_modificado.csv", sep="|")
```

```python
#Lista creada con tuplas con los cantones que estan registrados en la data
opciones =[(d,d)for d in datos_prefectos.CANTON_NOMBRE.unique()]
opciones.sort() # ordenados alfabeticamente
```

```python
#presento la lista
opciones
```

```python
#almaceno la data de los candidatos en la varible datos_candidatos
datos_candidatos = pd.read_spss("data/candidatos2019.sav")
datos_candidatos.shape #cantidad de filas y columnas
```

```python
#presenta los datos de la cabecera
datos_candidatos.head() 
```

```python
#opcion de cantones 
canton = widgets.Dropdown(options = opciones, value = "LOJA",description= 'Cantón',)
#boton de opciones
canton
```

```python
#consulta del canton seleccionado antes en el boton de opciones
datos_x_canton= datos_prefectos[datos_prefectos['CANTON_NOMBRE']==canton.value]
#presento cabecera de la consulta anterior
datos_x_canton.head()
```

```python
#muestra estadistica de los votos a los prefectos del canton 

datos_x_canton.groupby('CANDIDATO_NOMBRE')['VOTOS'].sum().sort_values().plot(kind="barh", width = 0.8)
```

```python
#muestra estadistica de los votos a los prefectos del canton 
datos_x_canton.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values(ascending=False).plot(kind="barh", width = 0.5)
```
